# Reproduce Auto Assign XHR Failure
This notebook sends a POST request to the local `/api/triage/auto_assign` endpoint and inspects the response to diagnose the server-side 500 error.

```python
import json
import urllib.request
import urllib.error

print('Ready to test auto_assign endpoint')
```

```python
url = 'http://127.0.0.1:8002/api/triage/auto_assign/852c3868-8b26-4029-97d6-1f63a681c3c9'
headers = {'Authorization': 'Bearer dev_token_test'}

print('URL:', url)
print('Headers:', headers)
```

```python
data = None
request = urllib.request.Request(url, method='POST', headers=headers)

try:
    with urllib.request.urlopen(request, timeout=20) as resp:
        body = resp.read().decode('utf-8', errors='replace')
        print('Response status:', resp.getcode())
        print('Response headers:', dict(resp.getheaders()))
        print('Response body:', body)
except urllib.error.HTTPError as exc:
    body = exc.read().decode('utf-8', errors='replace')
    print('HTTPError status:', exc.code)
    print('Response headers:', dict(exc.headers))
    print('Response body:', body)
except urllib.error.URLError as exc:
    print('URLError:', exc)
except Exception as exc:
    print('Unexpected error:', exc)
```

```python
overview_url = 'http://127.0.0.1:8002/api/triage/overview'
overview_request = urllib.request.Request(overview_url, method='GET', headers=headers)

try:
    with urllib.request.urlopen(overview_request, timeout=20) as resp:
        body = resp.read().decode('utf-8', errors='replace')
        print('\nOverview status:', resp.getcode())
        print('Overview headers:', dict(resp.getheaders()))
        print('Overview body:', body)
except urllib.error.HTTPError as exc:
    body = exc.read().decode('utf-8', errors='replace')
    print('\nOverview HTTPError status:', exc.code)
    print('Overview headers:', dict(exc.headers))
    print('Overview body:', body)
except urllib.error.URLError as exc:
    print('\nOverview URLError:', exc)
except Exception as exc:
    print('\nOverview Unexpected error:', exc)
```
